```python
regulons = regulons.drop([252], axis = 0).reset_index(drop = True)
```

```python
reg = []
for i in range(0, len(regulons.index)):
    if len(data[data["2)geneName"].str.lower() == regulons.iloc[i]["Regulon_Name"].lower()]) == 1:
        reg.append(data[data["2)geneName"].str.lower() == regulons.iloc[i]["Regulon_Name"].lower()])
```

```python
len(reg)
```

```python
data = pd.concat(reg).reset_index(drop = True)
```

```python
data.columns = list(pd.Series(data.columns).str[2:])
```

```python
data.columns
```

```python
data.columns = list(pd.Series(data.columns)[:-2]) + list(pd.Series(data.columns)[-2:].str[1:])
```

```python
data["geneName"] = data["geneName"].str.lower()
```

```python
regulons["Regulon_Name"] = regulons["Regulon_Name"].str.lower()
```

```python
data.columns = ['geneId', 'regulon_name', 'leftEndPos', 'rightEndPos', 'strand',
       'geneSynonyms', 'otherDbsGeneIds', 'productId', 'productName',
       'productSynonyms', 'otherDbsProductsIds ']
```

```python
regulons.columns = ['regulon_name', 'regulated_genes', 'regulated_operon', 'regulated_tu',
       'total_of_regulatory_interaction', 'total_of_binding_sites']
```

```python
regulons_data = pd.merge(regulons, data, on='regulon_name', how='left')
```

```python
regulons_data = regulons_data.dropna(subset=["otherDbsGeneIds"])
```
